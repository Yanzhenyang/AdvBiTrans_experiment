# Import Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Data fusion

```python
# 二分类数据平衡读取
benign = pd.read_csv("./Data/benign.csv")
benign = benign.iloc[:40000]
gafgyt_combo = pd.read_csv("./Data/gafgyt-combo.csv")
gafgyt_combo = gafgyt_combo.iloc[:10000]
gafgyt_junk = pd.read_csv("./Data/gafgyt-junk.csv")
gafgyt_junk = gafgyt_junk.iloc[:10000]
gafgyt_scan = pd.read_csv("./Data/gafgyt-scan.csv")
gafgyt_scan = gafgyt_scan.iloc[:10000]
gafgyt_tcp = pd.read_csv("./Data/gafgyt-tcp.csv")
gafgyt_tcp = gafgyt_tcp.iloc[:10000]
gafgyt_udp = pd.read_csv("./Data/gafgyt-udp.csv")
gafgyt_udp = gafgyt_udp.iloc[:10000]
mirai_ack = pd.read_csv("./Data/mirai-ack.csv")
mirai_ack = mirai_ack.iloc[:10000]
mirai_scan = pd.read_csv("./Data/mirai-scan.csv")
mirai_scan = mirai_scan.iloc[:10000]
mirai_syn = pd.read_csv("./Data/mirai-syn.csv")
mirai_syn = mirai_syn.iloc[:10000]
mirai_udp = pd.read_csv("./Data/mirai-udp.csv")
mirai_udp = mirai_udp .iloc[:10000]
mirai_udpplain = pd.read_csv("./Data/mirai-udpplain.csv")
mirai_udpplain = mirai_udpplain.iloc[:10000]
```

```python
# 更新gafgyt相关数据的label列
gafgyt_combo['label'] = 'gafgyt'
gafgyt_junk['label'] = 'gafgyt'
gafgyt_scan['label'] = 'gafgyt'
gafgyt_tcp['label'] = 'gafgyt'
gafgyt_udp['label'] = 'gafgyt'

# 更新mirai相关数据的label列
mirai_ack['label'] = 'mirai'
mirai_scan['label'] = 'mirai'
mirai_syn['label'] = 'mirai'
mirai_udp['label'] = 'mirai'
mirai_udpplain['label'] = 'mirai'

# 更新gafgyt相关数据的class列
gafgyt_combo['class'] = '2'
gafgyt_junk['class'] = '2'
gafgyt_scan['class'] = '2'
gafgyt_tcp['class'] = '2'
gafgyt_udp['class'] = '2'
```

```python
#数据融合
df = pd.concat([benign, gafgyt_combo , gafgyt_junk, gafgyt_scan, gafgyt_tcp, gafgyt_udp, mirai_ack, mirai_scan, mirai_syn, mirai_udp,mirai_udpplain])
```

```python
label_df = pd.DataFrame(df['label'].value_counts()).reset_index()
label_df

y = label_df['label'].values
mylabels = label_df['index'].values
print(y)

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = mylabels
sizes = y
explode = (0, 0, 0,)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots(figsize=(5, 5))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# 删除空值和无穷大值
df.replace([np.inf, -np.inf], np.nan, inplace= True)
df.info()
```

```python
# 统计每个特征中的空值
null_counts = df.isnull().sum()
print(null_counts)
```

```python
# 使用drop_duplicates()方法删除重复值
df = df.drop_duplicates()
```

```python
#删除缺失值的行
df = df.dropna()
```
