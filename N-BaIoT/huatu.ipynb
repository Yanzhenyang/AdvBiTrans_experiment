### accuracy

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

# 数据
accuracy_dnn_lstm = [99.95, 99.97, 99.89, 99.92, 99.92, 99.93, 99.96, 99.94, 99.94, 99.93]
accuracy_cnn2d_lstm = [99.95, 99.98, 99.90, 99.91, 99.94, 99.92, 99.96, 99.99, 99.92, 99.94]
accuracy_dcnn_lstm = [99.94, 99.94, 99.92, 99.96, 99.92, 99.89, 99.92, 99.92, 99.92, 99.97]
accuracy_rnn_lstm = [99.89, 99.90, 99.96, 99.92, 99.94, 99.88, 99.90, 99.92, 99.92, 99.95]

# 折数
folds = np.arange(1, 11)

# 函数，用于创建平滑的曲线
def make_smooth_line(x, y):
    xnew = np.linspace(min(x), max(x), 300)  # 创建300个点，以平滑曲线
    spl = make_interp_spline(x, y, k=3)  # 创建B样条曲线
    y_smooth = spl(xnew)
    return xnew, y_smooth

# 为每种算法生成平滑曲线
folds_smooth, accuracy_dnn_lstm_smooth = make_smooth_line(folds, accuracy_dnn_lstm)
_, accuracy_cnn2d_lstm_smooth = make_smooth_line(folds, accuracy_cnn2d_lstm)
_, accuracy_dcnn_lstm_smooth = make_smooth_line(folds, accuracy_dcnn_lstm)
_, accuracy_rnn_lstm_smooth = make_smooth_line(folds, accuracy_rnn_lstm)

# 创建图表，仅在每个折点位标记
plt.figure(figsize=(7, 4))

# 绘制平滑曲线和特定标记
plt.plot(folds_smooth, accuracy_dnn_lstm_smooth, color='black', label='DNN-LSTM')
plt.scatter(folds, accuracy_dnn_lstm, marker='o', color='black')
plt.plot(folds_smooth, accuracy_cnn2d_lstm_smooth, color='red', label='CNN2D+LSTM')
plt.scatter(folds, accuracy_cnn2d_lstm, marker='s', color='red')
plt.plot(folds_smooth, accuracy_dcnn_lstm_smooth, color='blue', label='DCNN+LSTM')
plt.scatter(folds, accuracy_dcnn_lstm, marker='^', color='blue')
plt.plot(folds_smooth, accuracy_rnn_lstm_smooth, color='green', label='RNN-LSTM')
plt.scatter(folds, accuracy_rnn_lstm, marker='v', color='green')

# 添加标题和标签
plt.title('10-Fold Cross Validation Accuracy',fontsize=12, fontweight='bold')
plt.xlabel('Folds',fontsize=12, fontweight='bold')
plt.ylabel('Accuracy (%)',fontsize=12, fontweight='bold')

# 添加图例
plt.legend()

# 添加网格
plt.grid(True)

# 保存新图像
plt.savefig("./Picture/Mult_Feature_select_PPAC/10_ACC.png")
# 显示图表
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Metric names and their values for each algorithm (in percentages)
metrics = ['TNR', 'MCC', 'NPV']
BiTrans_values = [99.96, 99.88, 99.96]
AdvBiTrans_values = [99.97, 99.91, 99.96]

# Calculate the difference to stack only the incremental part
AdvBiTrans_increment = [adv - bitrans for adv, bitrans in zip(AdvBiTrans_values, BiTrans_values)]

# Setting the positions and width for the bars
group_margin = 0.25
pos = np.arange(len(metrics))
bar_width = 0.1  # Original bar width
space_for_gap = 0.01  # Space for gap between the main bar and the increment

# Colors for the two algorithms
colors = ['blueviolet', 'cornflowerblue']

# Plotting
plt.figure(figsize=(6, 6))
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

# Plot each set of bars for 'BiTrans' and the incremental 'AdvBiTrans'
bars_bi_trans = plt.bar(pos, BiTrans_values, bar_width, label='BiTrans', color=colors[0])
# Add a small space by subtracting from the bottom value
bars_adv_increment = plt.bar(pos, AdvBiTrans_increment, bar_width, bottom=np.array(BiTrans_values) + space_for_gap, 
                             label='AdvBiTrans Increment', color=colors[1])

# Function to annotate the incremental value on top of the stack
def annotate_increment(bars, increments, fontsize):
    for bar, increment in zip(bars, increments):
        if increment > 0:  # Only annotate positive increments
            plt.annotate(f'+{increment:.2f}',
                         xy=(bar.get_x() + bar.get_width() / 2, bar.get_height() + space_for_gap),
                         xytext=(0, 3),  # 3 points vertical offset
                         textcoords="offset points",
                         ha='center', va='bottom',
                         rotation=90, fontsize=fontsize)

# Adding annotation for the incremental values
annotate_increment(bars_adv_increment, AdvBiTrans_increment, fontsize=8)

# Setting bold labels and title
plt.xlabel('Metrics', fontweight='bold', fontsize=14)
plt.ylabel('Percentage (%)', fontweight='bold', fontsize=10)
plt.title('Performance Increment of AdvBiTrans over BiTrans', fontweight='bold', fontsize=14)
plt.xticks(pos, metrics, fontweight='bold', fontsize=10)
plt.legend()
plt.ylim(99.80, 100)  # Adjust y-axis limits for visibility

# Save the updated plot
#plt.savefig("/mnt/data/Stacked_Increment_Performance_with_Gap.png")

# Display the plot
plt.show()
```
